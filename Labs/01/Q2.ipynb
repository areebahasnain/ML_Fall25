```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# B. Load Dataset
df = pd.read_csv("cancer_patient.csv")

# C. Find out the unique category in target variable
df['Level'].unique()
```

```python
# D. Dataset Overview
print("Total No. of Columns: ", len(df.columns))
display(df.columns)
print("Shape:", df.shape)

display(df.head(50))    # first 50 rows
display(df.tail(30))    # last 30 rows

print(df.info())        # memory usage, datatypes
```

```python
# E. No. of columns & samples
display(len(df.columns))
display(len(df))
```

```python
# F. Target Variable Distribution (Numerical & Graphical Visualization)
print("Unique categories in target variable:\n", df['Level'].value_counts())

plt.figure(figsize=(6,4))
sns.countplot(x='Level', hue='Level', data=df, palette="Set2", legend=False)
plt.title("Distribution of Target Variable")
plt.xlabel("Target Categories")
plt.ylabel("Number of Records")
plt.show()
```

```python
# G. Extract 550â€“900 samples and save in a new csv
subset_loc = df.loc[550:900, 'Gender':'Coughing of Blood']
subset_loc.to_csv("subset_loc.csv", index=False)

subset_loc.describe()
```

```python
# H. Extract Last 20 Records (Features: Shortness of Breath â†’ Frequent cold) and save in a new csv

# get column index positions
start_col = df.columns.get_loc('Shortness of Breath')
end_col = df.columns.get_loc('Frequent Cold') + 1 # +1 because iloc is exclusive at the end

# extract last 20 rows and 5 features using iloc
subset_iloc = df.iloc[-20:, start_col:end_col]

# save dataset to new csv
subset_iloc.to_csv("subset_iloc.csv", index=False)

# statistical analysis
subset_iloc.describe()
```

```python
# I. Extract Records [110,220,360,440,656,778,202,889] & Find Average Alcohol Use
records = df.loc[[110, 220, 360, 440, 656, 778, 202, 889], :]
display(records)

avg_alcohol = records['Alcohol use'].mean()
print("Average Alcohol Use: ", avg_alcohol)
```
