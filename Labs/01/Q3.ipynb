```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cancer_patient.csv")
```

```python
# A. Perform Graphical Visualization
plt.figure(figsize=(10,6))
df['Level'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Level Distribution")
plt.xlabel("Categories")
plt.ylabel("Count")
plt.show()

df.hist(figsize=(11, 13), bins=30, color='plum')
plt.suptitle("Histograms of Numerical Features", fontsize=16)
plt.show()
```

```python
# B. Correlation Graph
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="Pastel2", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
# C. Distribution of Age and Frequent Cold
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x='Frequent Cold', data=df, hue='Frequent Cold', palette="Pastel2", legend=False)
plt.title("Distribution of Frequent Cold")
plt.xlabel("Frequent Cold (Yes/No)")
plt.ylabel("Count")
plt.show()
```

```python
# D. Outliers in Obesity Feature
plt.figure(figsize=(6,3))
sns.boxplot(x=df['Obesity'], color="palegreen")
plt.title("Boxplot of Obesity Feature")
plt.xlabel("Obesity")
plt.show()
```
